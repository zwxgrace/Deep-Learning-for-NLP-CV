```python
data = open('LSTM.txt').read()
# 移除换行符
data = data.replace('\n','')
data = data.replace('\\','')
```

## 进行数据预处理 将文字序列数据转化为可用于LSTM输入的数据

```python
#字符去重处理
letters = set(list(data))
print(len(letters)) 
```

```python
#建立字典用数字进行索引
int_to_char = {a:b for a,b in enumerate(letters)}
print(int_to_char)
```

```python
# 一般将字母转换成对应的数值，在进行onehot编码
char_to_int = {a:b for b,a in enumerate(letters)}
print(char_to_int)
```

```python
# 告诉计算机要用20个数据预测第21个
time_step = 20
```

```python
# 进行字符串批量处理
from tensorflow.keras.utils import to_categorical
```

```python
# 滑动窗口提取数据
def extract_data(data,slide): # slide是time_step
    x = []
    y = []
    for i in range(len(data)-slide):
        x.append([a for a in data[i:i+slide]])  # 取左不取右
        y.append(data[i+slide])
    return x,y
```

```python
extract_data(data,time_step)
```

```python
# 字符到字符串的批量转化
def char_to_int_data(x,y,char_to_int):
    x_to_int = []
    y_to_int = []
    for i in range(len(x)):
        x_to_int.append([char_to_int[a] for a in x[i]])
        y_to_int.append([char_to_int[a] for a in y[i]])
    return x_to_int,y_to_int
```

```python
# 实现输入字符文章的批量处理，输入整个字符、滑动窗口大小、转化字典
def data_preprocessing(data,slide,num_letters, char_to_int):
    char_data = extract_data(data,slide)
    int_data = char_to_int_data(char_data[0],char_data[1],char_to_int)
    
    Input = int_data[0]  #定义输入值是数值化后的
    output = list(np.array(int_data[1]).flatten())
    Input_reshaped = np.array(Input).reshape(len(Input),slide) # reshape成矩阵形式
    
    new = np.random.randint(0,10,size = [Input_reshaped.shape[0],Input_reshaped.shape[1],num_letters])
    for i in range(Input_reshaped.shape[0]):
        for j in range(Input_reshaped.shape[1]):
            new[i,j,:] = to_categorical(Input_reshaped[i,j],num_classes = num_letters)  #X已经变成onehot格式了
    return new,output
```

```python
X,y = data_preprocessing(data,time_step,61, char_to_int)
```

## 查看文字数据预处理后的数据结构 并进行数据分离操作

```python
X.shape 
# 1189个矩阵 ， 每个矩阵是20*41的
# 每一个字母相当于是一个41维度的onehot矩阵，
```

```python
len(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
# split the data
X_train, X_test,y_train,y_test = train_test_split(X,y,test_size =0.2, random_state = 100)
print(X_train.shape,len(y_train))
```

```python
y_train_catego = to_categorical(y_train,61)  #把y_train进行一个41维度的onehot编码
```

```python
print(y_train_catego.shape)
y_train_catego
```

## 建立模型
## 针对字符串输入（ “   ”）预测对应的后续字符

```python
from keras.models import Sequential
from keras.layers import Dense,LSTM
```

```python
model = Sequential()
model.add(LSTM(units = 20,input_shape = (X_train.shape[1],X_train.shape[2]),activation= 'relu'))
model.add(Dense(units = 61,activation= 'softmax'))

model.compile(optimizer = 'adam', loss = 'categorical_crossentropy',metrics = ['accuracy'])
model.summary()          
```

```python
model.fit(X_train,y_train_catego,epochs = 15)
# 进行多次训练 提高准确度
```

```python
# train the model
#model.fit(X_train,y_train,batch_size = 200,epochs = 5)
# 样本比较大的时候可以分批处理，这里是200个一批处理了5批次
model.fit(X_train,y_train_catego,batch_size = 1000,epochs = 15)
```

```python
# predict
y_train_pred = np.argmax(model.predict(X_train),axis =1)
```
